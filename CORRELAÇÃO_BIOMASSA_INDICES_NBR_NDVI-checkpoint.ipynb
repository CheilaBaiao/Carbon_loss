```python
x = df['before']
y = df['after']

# Ajusta uma regressão de potência (y = a * x^b) aos dados
coefficients = np.polyfit(np.log(x), np.log(y), 1)
a = np.exp(coefficients[1])
b = coefficients[0]

# Calcule o coeficiente de determinação (R²)
y_predicted = a * (x ** b)
ssr = np.sum((y_predicted - np.mean(y)) ** 2)
sst = np.sum((y - np.mean(y)) ** 2)
r_squared = 1 - (ssr / sst)

# Crie um array de valores de x para a curva de ajuste
x_fit = np.linspace(min(x), max(x), 100)

# Calcule os valores de y ajustados
y_fit = a * (x_fit ** b)

# Plote os dados e a curva de ajuste com R² na legenda
plt.scatter(x, y, label=f'Dados\nR²: {r_squared:.4f}')
plt.plot(x_fit, y_fit, 'r-', label=f'Regressão de Potência: y = {a:.2f} * x^{b:.2f}')
plt.xlabel('Variável "Before"')
plt.ylabel('Variável "After"')
plt.legend()
plt.show()
```
