```python
import textwrap
def wrap_labels(ax, width, break_long_words=False):
    labels = []
    for label in ax.get_xticklabels():
        text = label.get_text()
        labels.append(textwrap.fill(text, width=width,
                      break_long_words=break_long_words))
    ax.set_xticklabels(labels, rotation=0)
```

```python
sns.set(style="darkgrid")
axO= sns.boxplot(data=df, x='LULC', y='Carbon storage', showmeans = True, meanprops={"marker":"o",
                       "markerfacecolor":"white", 
                       "markeredgecolor":"black",
                      "markersize":"8"}, order=['Pasture', 'HAG', 'Eucalyptus', 'SSF secondary', 'SSF Old-growth', 'ODF Secondary', 'ODF Old-growth', 'OMF secondary', 'OMF Old-growth'], showfliers = False)
axO.set(ylabel=None)
axO.set(xlabel=None)
wrap_labels(axO, 10)
axO.figure.set_size_inches(10,6)

axO.set_title('(A) Carbon loss per LULC (Mg C)', fontsize=16)
plt.show
fig = axO
plt.savefig ('carbon_lulc.tif')
```
